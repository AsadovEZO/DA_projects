Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Для изучения данных будем использовать функцию inspect.

```python
def inspect(df):
    display(df.sample(10, random_state=1)) 
    df.info() 
    print ('Количество дубликатов -', df.duplicated().sum())
    numeric_columns=df.describe().columns
    display(df.describe())
    display()
    df.columns = [x.lower().replace(' ', '_') for x in df.columns.values]
    for column_name in df.columns:
        if 'datetime' in column_name:
            df[column_name] = pd.to_datetime(df[column_name])
        else:
            display(df[column_name].value_counts())
            print(50*'*')
    return df
```

### Преобразование данных в нужные типы
Столбцы с названием игры, платформой и ее жанром имеют формат object, что соответствует их содержанию.  
year_of_release содержит год выпуска игры. Для экономии памяти будем использовать для этого столбца тип данных int16. Но перевести тип данных для этого столбца мы сможем только после заполнения пустых значений (в следующем подразделе).  
Продажи для той же экономии памяти можно перевести в формат float32 (при переводе в тип float16 теряется определенная точность данных).  
Оценки user_score переведем в float32. Однако для этого нам также потребуется обработать пропуски, а также значения текстовые значения "tbd" (в следующем подразделе).

```python
df = df.astype({'na_sales': 'float32', 'eu_sales': 'float32', 'jp_sales': 'float32', 
                'other_sales': 'float32', 'critic_score': 'float32'})
```

### Обработка пропусков  
Проверим пустые значения в столбце name.

```python
df[df['name'].isna()]
```

Игры, у которых отсутствует название и жанр, к сожалению, не помогут в проведении исследования. Самостоятельно по имеющимся данным мы не сможем заполнить эти пропуски.  
Удаление 2 строк из массива не повлияет на результаты исследования. 

```python
df = df.dropna(subset = ['name']).reset_index(drop = True)
```

Далее проверим пустые значения в столбце с годом выпуска игры.  
После обработки пропусков в этом столбце, заменим его тип на int16 (о чем писали ранее).

```python
df[df['year_of_release'].isna()]
```

Теоретически, мы можем заполнить пропуски на основе данных из интернета, но так как пропусков слишком много, это неэффективно.  
Попробуем заполнить пропуски на основе имеющихся в таблице данных. Одни и те же игры могли выходить на разных платформах, и данный метод заполнения будет чуть более приближен к реальным значениям, нежели чем заполнение медианами по платформе. 

```python
df['year_of_release'] = df['year_of_release'].fillna(df.groupby('name')['year_of_release'].transform('median'))
len(df[df['year_of_release'].isna()])
```

В таблице осталось 146 строк с пропусками в столбце year_of_release.  
Остальные пропуски заполним медианными значениями на основе платформы. Оставшихся пропусков меньше, чем 1% от общего количества данных, поэтому подобное заполнение не должно сильно исказить результаты.  
Предпочтительно не удалить эти пропуски, а заполнить значениями, т.к. в противном случае мы потеряем много данных, находящихся в других столбцах таблицы.

```python
df['year_of_release'] = df['year_of_release'].fillna(df.groupby('platform')['year_of_release'].transform('median'))
```

Переведем формат столбца в int16.

```python
df['year_of_release'] = df['year_of_release'].astype('int16')
```

Рейтинг ESRB используется только в Северной Америке. Отсутствие данных для других регионов - не ошибка. Проверим, какое кол-во игр таблицы не имеют рейтинга.

```python
print('Количество игр без рейтинга ESRB -', len(df[df['rating'].isna()]), 
      '({:.2%} от общего количества игр).'.format(len(df[df['rating'].isna()])/len(df['rating'])))
```

Заменим пропуски в рейтинге на 'no rating' для дальнейшего анализа.

```python
df['rating'] = df['rating'].fillna('no rating')
```

Остались пропуски в столбцах critic_score, user_score и в rating (пропуски с оценками подтягивать на основе оценок игры на других платформах некорректно, т.к. у одной игры для разныхх платформ могут быть разные оценки). Эти пропуски заполнить самостоятельно мы не сможем, и их количество слишком велико, чтобы удалять эти данные.  
Замена пропусков на нули для корректной смены типа данных не целесообразна, так как это негативно повлияет на метрики, которые будут рассчитываться в дальнейшем.  

Пропуски в столбцах name, genre, year_of_release похожи на механические ошибки (при обработке/переносе данных).  
Пропуски в столбцах с оценками и рейтингом не похожи на случайные. Одна из возможных причин - ряд игр выходили еще до того, как данные оценки и рейтинги были введены.  


### Проверка данных  
В столбце с пользовательскими оценками user_score у ряда строк вместо оценки указано значение "tbd". Эта аббревиатура означает "To Be Determined" - "Будет определено". То есть это те игры, у которых еще не опреден пользовательский рейтинг (допускаю, что по причине отсутствия требуемого количества пользовательских оценок).
Проверим, у игр какого года отсутствуют пользовательские оценки.

```python
df.query('user_score == "tbd"').pivot_table(index = 'year_of_release', values = 'name', aggfunc = 'count')
```

Да, данный рейтинг указан не только у игр, вышедших за 2016 год (по которым не могло набраться нужное количество пользовательских оценок в связи с недавним выходом), но и у игр 10-20-ти летней давности.  
Так как это игры, по которым, по факту, не был определен пользовательский рейтинг, заменим эту оценку на NaN (для дальнейшей замены типа данных на float32). 

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'] = df['user_score'].astype('float32')
```

Проверим данные последнего столбца - rating.  

```python
df['rating'].value_counts()
```

В данных очень мало игр с рейтингом EC (Для детей младшего возраста), K-A (рейтинг, который в 1998 году был переименован в Е), RP ("Рейтинг ожидается"), а также AO (Adults Only 18+).  
Чтобы не отвлекаться на эти сверхмалые категории рейтинга, обработаем их.  
EC объединим с Е (игры для всех), K-A заменим на Е, RP заменим на NaN, одну игру с рейтингом AO объединим с группой рейтинга М (От 17 лет).

```python
df['rating'] = df['rating'].replace('EC', 'E')
df['rating'] = df['rating'].replace('K-A', 'E')
df['rating'] = df['rating'].replace('RP', np.nan)
df['rating'] = df['rating'].replace('AO', 'M')
```

Далее проверим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Для этого построим матрицу диаграмм рассеяния, используя функцию для автоматического создания матриц.  
Для построения матрицы будем использовать только данные, по которым есть оценки критиков и пользователей.  

```python
def corr_mtx(pt):
    print ('\033[1m' + 'Матрица диаграмм рассеяния для платформы ' + pt + '\033[0m')
    pd.plotting.scatter_matrix(df_actual[df_actual['platform'] == pt].query
                               ('critic_score > 0 and user_score > 0')[
                                   ['overall_sales', 'critic_score', 'user_score']], figsize=(9, 9)) 
    plt.show()
    print ('\033[1m' + 'Коэффициенты корреляции между продажами и оценками ' + pt + '\033[0m')
    print(df_actual[df_actual['platform'] == pt].query('critic_score > 0 and user_score > 0')[
                                   ['overall_sales', 'critic_score', 'user_score']].corr()) 
```

```python
corr_mtx('PS4')
```

Корреляция между продажами на PS4 и оценками пользователей отсутствует (о чем говорит коэффициент -0,03).  
Оценки критиков, в этом случае, являются куда более приоритетными, и, с учетом коэффициента корреляции 0,41, можно говорить об определенной взаимосвязи между продажами и оценками критиков.  
Также, что интересно, наблюдается взаимосвязь между оценками критиков и пользователей, которая сильнее, чем взаимосвязь между продажами и оценками пользователей (0,56).  
Это же наблюдается на матрице рассеяния.

## Проверка гипотезы 
В рамках последнего раздела будут проверены 2 гипотезы:  
1) Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;  
2) Средние пользовательские рейтинги жанров Action и Sports разные.  

Для проверки гипотез возьмем вероятность ошибочно отклонить нулевую гипотезу в размере 5%. 

### Проверка гипотезы №1
Первая проверяемая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
Сформулируем нулевую и альтернативную гипотезы:  
H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые (по формулировке соответствует нашей гипотезе).  
H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются.  

Для проверки нулевой гипотезы требуется посчитать вероятность получить наблюдаемое на выборке значение при условии, что нулевая гипотеза верна. В зависимости от полученной вероятности сделаем вывод — требуется ли отвергать H₀.  

Для проверки сделаем t-тест.  
Но сначала проверим количество данных для понимания, насколько целесообразно применять данный тип теста без более детальных проверок. 

```python
print('Количетсво игр с пользовательским рейтингом на платформе PC:', 
      len(df_actual[(df_actual['platform'] == 'PC') & (df_actual['user_score'].isna() == False)]))
print('Количетсво игр с пользовательским рейтингом на платформе XBox One:', 
      len(df_actual[(df_actual['platform'] == 'XOne') & (df_actual['user_score'].isna() == False)]))
```

Данного количества наблюдений вполне достаточно для проведения t-теста. Так как пользовательские оценки не могут быть более 10, проверять данные на наличие выбросов не требуется. 

```python
results1 = st.ttest_ind(df_actual[(df_actual['platform'] == 'PC') & (df_actual['user_score'].isna() == False)]['user_score'], 
df_actual[(df_actual['platform'] == 'XOne') & (df_actual['user_score'].isna() == False)]['user_score'])
alpha = 0.05
print('p-значение:', results1.pvalue)
if results1.pvalue < alpha: # сравниваем получившееся p-значение с заданным уровнем статистической значимости:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
print('Средние пользовательские рейтинги по платформе Xbox One', 
      df_actual[(df_actual['platform'] == 'XOne') & (df_actual['user_score'].isna() == False)]['user_score'].mean())
print('Средние пользовательские рейтинги по платформе PC', 
      df_actual[(df_actual['platform'] == 'PC') & (df_actual['user_score'].isna() == False)]['user_score'].mean())
```

Нам не удалось отвергнуть нулевую гипотезу (которая соответствует первой). С очень высокой долей вероятности (p-значение 0.625) можно говорить о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, что можно увидеть и по средним значениям выше.

### Проверка гипотезы №2


Гипотеза 2: Средние пользовательские рейтинги жанров Action и Sports разные.  
H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые.  
H₁: Средние пользовательские рейтинги жанров Action и Sports различаются.

```python
print('Количетсво игр с пользовательским рейтингом для игр жанра Action:', 
      len(df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'].isna() == False)]))
print('Количетсво игр с пользовательским рейтингом для игр жанра Sports:', 
      len(df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'].isna() == False)]))
```

Игр с жанром Action значительно больше, но данных все равно достаточно для проведения теста. 

```python
results2 = st.ttest_ind(df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'].isna() == False)]['user_score'], 
df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'].isna() == False)]['user_score'])
alpha = 0.05
print('p-значение:', results2.pvalue)
if results2.pvalue < alpha: # сравниваем получившееся p-значение с заданным уровнем статистической значимости:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Р-значение по итогам проведения теста оказалось критически мало. У нас нет оснований принять нулевую гипотезу.  
Проверим визуально, как отличаются средние пользовательские оценки для игр этих двух жанров. 

```python
df_actual[(df_actual['genre'] == 'Action') & 
          (df_actual['user_score'].isna() == False)]['user_score'].hist(alpha=0.7, bins=50, figsize=(15, 10))
df_actual[(df_actual['genre'] == 'Sports') & 
          (df_actual['user_score'].isna() == False)]['user_score'].hist(alpha=0.6, bins=50, figsize=(15, 10))
plt.legend(['Action', 'Sports'])
plt.title('Гистограмма распределения пользовательских оценок видеоигр жанров Action и Sports за период 2012-2016 гг'
          , fontsize=16)
plt.xlabel('Пользовательские оценки', fontsize=14)
plt.ylabel('Количество наблюдений', fontsize=14)
plt.show()
```

```python
print('Средние пользовательские рейтинги по жанру Action', 
      df_actual[(df_actual['genre'] == 'Action') & (df_actual['user_score'].isna() == False)]['user_score'].mean())
print('Средние пользовательские рейтинги по жанру Sports', 
      df_actual[(df_actual['genre'] == 'Sports') & (df_actual['user_score'].isna() == False)]['user_score'].mean())
```
