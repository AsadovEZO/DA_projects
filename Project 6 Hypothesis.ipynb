## Часть 1. Приоритизация гипотез
В рамках данной части проекта нам требуется приоретизировать гипотезы.  
На вход нам было предоставлено 9 гипотез для улучшения метрик бизнеса. По этим гипотезам уже произвели требуемые для приоритезации расчеты метрик. Наша задача - выстроить приоритеты для данных гипотез. То есть оценить, какие гипотезы принесут максимальное ожидаемое улучшение метрик бизнеса.  
Для выполнения задачи будут применены фреймворки ICE и RICE. Они позволят произвести оценку имеющихся гипотез на основе таких метрик, как охват пользователей, степень влияния изменения на охваченных пользователей, степерь уверенности в оценках других параметров и оценки сложности проверки гипотез.

### 1.1 Импорт и изучение данных
Импортируем и изучим предоставленные данные

```python
df_hypothesis = pd.read_csv('/datasets/hypothesis.csv')
df_hypothesis
```

В файле представлены гипотезы, а также показатели охвата пользователей reach, степень влияния изменения на охваченных пользователей impact, степерь уверенности в оценках других параметров confidence и оценка сложности проверки гипотез efforts.

### 1.2 Расчет ICE для гипотез
Первый фреймворк, с помощью которого будем приоритезировать имеющиеся гипотезы, - ICE.  
Для расчета ICE потребуются показатели impact, confidence и efforts.  
Добавим расчетный столбец с фреймворком в конец таблицы, отсортируем таблицу по убыванию ICE. 

```python
df_hypothesis['ICE'] = df_hypothesis['Impact'] * df_hypothesis['Confidence'] / df_hypothesis['Efforts']
df_hypothesis.sort_values(by='ICE', ascending=False)
```

Согласно фреймворку ICE, лидером среди гипотез будет гипотеза №9 - "Запустить акцию, дающую скидку на товар в день рождения". ICE этой гипотезы составляет 16,20.  
Следующими идут гипотезы №1 "Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей" (ICE 13.33) и №8 "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок" (ICE 11.20).

### 1.3 Расчет RICE для гипотез
Второй фреймворк для приоритезации гипотез - RICE.  
При расчете RICE будет задействована оценка охвата пользователей.  
Также добавим расчетный столбец с фреймворком в конец таблицы, отсортируем таблицу по убыванию RICE. 

```python
df_hypothesis['RICE'] = (df_hypothesis['Reach'] * df_hypothesis['Impact'] 
                         * df_hypothesis['Confidence']) / df_hypothesis['Efforts']
df_hypothesis.sort_values(by='RICE', ascending=False)
```

При добавлении в расчет показателя reach, приоритет гипотез смещается. Безоговорочным лидером стала гипотеза №8 - "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок". Ее RICE составил 112.00, что в два раза больше, чем у ближайшего конкурента.  
Добавим цветовую окраску ячеек для наглядности.

```python
df_hypothesis['RICE'] = (df_hypothesis['Reach'] * df_hypothesis['Impact'] 
                         * df_hypothesis['Confidence']) / df_hypothesis['Efforts']
df_hypothesis.sort_values(by='RICE', 
                          ascending=False).style.background_gradient(subset=['ICE','RICE'], cmap='magma')
```

### 1.4 Вывод
При приоритезации гипотез без учета показателя охвата пользователей лидером была гипотеза №1 - "Запустить акцию, дающую скидку на товар в день рождения". Однако при учете этого показателя приоритеты гипотез сместились, и лидером стала гипотеза №3 - "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок".  
Это связано с тем, что гипотеза №1 имеет минимальный охват пользователей, в то время как гипотеза №3 имеет наоборот, наибольшее число пользователей, которых затронет изменение.  
Дополнительными преимуществами гипотезы №3 можно назвать хорошее влияние предлагаемого в ней изменения (7 из 10) и степень уверенности в оценках других параметров (8 из 10), с учетом среднего показателя сложности проверки гипотезы (5 из 10).

### 2.2 Расчет, анализ и визуализация кумулятивных метрик
Для расчета кумулятивных метрик создадим отдельный датафрейм cumulativeData.
#### 2.2.1 Собираем кумулятивные метрики
Для сборки кумулятивных метрик нам потребуется массив уникальных пар значений дат и групп. Сохраним его в переменной datesGroups

```python
datesGroups = df_orders[['date', 'group']].drop_duplicates()
```

Агрегируем кумулятивные значения по дням для заказов и посещений. 

```python
ordersAggregated = datesGroups.apply(
    lambda x: df_orders[
        np.logical_and(
            df_orders['date'] <= x['date'], df_orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transactionid': 'nunique',
            'visitorid': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

visitorsAggregated = datesGroups.apply(
    lambda x: df_visitors[
        np.logical_and(
            df_visitors['date'] <= x['date'], df_visitors['group'] == x['group']
        )
    ].agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}),
    axis=1,
).sort_values(by=['date', 'group'])
```

Собираем полученные кумулятивные данные в общую таблицу cumulativeData для дальнейшего анализа кумулятивных показателей.

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
cumulativeData
```

#### 2.2.2 Кумулятивная выручка по группам
Для анализа кумулятивной выручки по группам разделим группы А и В на разные массивы, после чего построим графики по дням.

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# Увеличиваем размер области построения
plt.figure(figsize=(10, 5))
# Строим график выручки группы А
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
# Строим график выручки группы B
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.title('Кумулятивная выручка групп А и В по дням за август 2019', fontsize=16)
plt.legend() 
plt.show()
```

По построенному графику кумулятивных выручек групп А и В можно сделать следующие выводы:
1) Кумулятивная выручка растет у обоих групп, что может служить индикатором корректности посчитанных данных;  
2) В начале эксперимента выручка обоих групп была примерно на одном уровне. На пятый день выручка группы В вырвалась вперед относительно выручки группы А, после чего на протяжении всего остального эксперимента была выше. Единственная точка выравнивания кумулятивных выручек - 13.08.2019, после чего рост выручки группы А замедлился в сравнении со второй группой;  
3) 18.08.2019 наблюдается резкий всплеск выручки группы В. Повод для проверки. Вероятно, это выброс;  
4) На основании кумулятивной выручки групп без очистки данных от выбросов можно сделать вывод, что группа В приносит больше выручки.  

#### 2.2.3 Кумулятивный средний чек по группам

```python
plt.figure(figsize=(10, 5))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.title('График динамики кумулятивного среднего чека А и В по дням за август 2019', fontsize=16)
plt.legend() 
plt.show()
```

По графику динамики кумулятивного среднего чека можно сделать следующие выводы:

Рост показателя в первой половине августа неравномерный. 18.08 наблюдается резкий скачок среднего чека, что совпадает со скачком кумулятивной выручки на графике. С 19.08 наблюдается систематичное уменьшение среднего чека. Предположительно, его падение продолжится вполть до достижения своего среднего значения;  
3) Для оформления выводов по этому графику требуются дополнительные наблюдения для достижения группой В равномерного графика среднего чека, или же очистка данных от выбросов.

### 2.4 Проверка статистической значимости различий между группами
Для проверки статистической значимости между двумя рассматриваемыми группами будет проверяться теория о том, что различий между группами нет. Альтернативная теория будет говорить, что различия между группами есть.  
p-value для всех тестов будет приниматься за 0,05.  
Для проверки гипотез будет использоваться U-критерий Манна-Уитни.  
Для "сырых" и "очищеных" данных проверим статистическую значимость различий в конверсии и в среднем чеке.   
#### 2.4.1 Проверка статистической значимости различий между группами по "сырым" данным 
##### 2.4.1.1 Проверка статистической значимости различий в конверсии между группами по "сырым" данным 
Для сравнения кумулятивной конверсии и среднего чека между двумя группами сначала выделим кумулятивные метрики по группам в отдельные датафреймы, а затем объединим их в единую общую таблицу.  
Для удобства выведем кумулятивную конверсию по группам, полученную ранее.


```python
visitorsADaily = df_visitors[df_visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']


visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = df_visitors[df_visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']


ordersADaily = (
    df_orders[df_orders['group'] == 'A'][['date', 'transactionid', 'visitorid', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionid': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    df_orders[df_orders['group'] == 'B'][['date', 'transactionid', 'visitorid', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionid': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)

print(data.head(5))
```

Далее выделим заказы и посещения для обеих групп в разные переменные, и проверим p-value по тесту Манна-Уитни.

Проверяемые гипотезы:
- Нулевая гипотеза H₀: кумулятивная конверсия групп А и В равна. Различий между группами нет.  
- Альтернативная гипотеза H₁: кумулятивная конверсия групп А и В не равны. Между группами есть статистически значимые различия.    

```python
ordersByUsersA = (
    df_orders[df_orders['group'] == 'A']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    df_orders[df_orders['group'] == 'B']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders'] 

sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
alpha = 0.05 # задаем уровень статистической значимости для текущего и последующих тестов:
print('p-value по итогам теста составляет {0:.5f}'.format(st.mannwhitneyu(sampleA, sampleB)[1]))
if st.mannwhitneyu(sampleA, sampleB)[1] < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
print('Относительный прирост конверсии группы В по отношению к группе А составил {0:.1%}'.format(sampleB.mean() / sampleA.mean() - 1))
```

p-value по итогам теста оказался ниже ранее установленного уровня в 5%, нулевая гипотеза не принимается - между группами есть статистически значимые различия.  
##### 2.4.1.2 Проверка статистической значимости различий в среднем чеке между группами по "сырым" данным 


Проверяемые гипотезы:
- Нулевая гипотеза H₀: кумулятивные средние чеки групп А и В равны.    
- Альтернативная гипотеза H₁: кумулятивные средние чеки групп А и В не равны. Между средними чеками групп есть статистически значимые различия. 

```python
print('p-value по итогам теста составляет {0:.5f}'.format(st.mannwhitneyu(df_orders[df_orders['group']=='A']['revenue'], 
                                       df_orders[df_orders['group']=='B']['revenue'])[1]))
if st.mannwhitneyu(df_orders[df_orders['group']=='A']['revenue'], 
                                       df_orders[df_orders['group']=='B']['revenue'])[1] < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
print('Относительный прирост среднего чека группы В по отношению к группе А составил {0:.1%}'
      .format(df_orders[df_orders['group']=='B']['revenue'].mean()/
                       df_orders[df_orders['group']=='A']['revenue'].mean()-1)) 
```

Проверка дала противоречивые результаты. С очень высокой вероятностью (более 70%) нулевая гипотеза подтверждается - то есть различий в среднем чеке между группами А и В нет, однако относительный прирост среднего чека составил почти 26%.  
Такой прирост, вероятнее всего, связан со скачком 18.08.2019, о чем писали ранее.  
Проверим это далее, в очищеных от выбросов данных.
