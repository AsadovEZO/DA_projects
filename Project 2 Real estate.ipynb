### Обработка пропусков.  
Пропуски есть почти во всех столбцах таблицы. Проверим, в каких столбцах отсутствуют данные, и что с ними потребуется делать для решения поставленной задачи.
В рамках обработки пропусков мы не будем удалять строки таблицы, т.к. вместе с пропусками могут быть удалены данные, которые могут повлиять на результаты анализа.

В столбце ceiling_height (высота потолков (м)) значения отсутствуют почти в 40% случаев (у 9195 строк отсутствуют данные в этом столбце). Заполнение данных для этого столбца вручную (даже на основе нескольких коррелирующих величин) может негативно повлиять на итоговые данные, сильно сместив их к усредненным значениям. Так как данная характеристика не требуется нам для проведения анализа, мы не ставим себе целью заполнение пропусков для столбца ceiling_height. Пропустим заполнение данных для этого столбца.

В столбце floors_total (всего этажей в доме) 86 пропусков (что составляет 0,36%). Отсутствие данных в этом столбце не повлияет на результаты исследования.

living_area (жилая площадь в квадратных метрах (м²)) - отсутствует 1903 значения (8%).  
Наличие данных в этом столбце для нас критично.  
Заполним эти данные на основании общей площади квартиры.  
Корреляция этих величин может быть очевидна, но проверим ее.

```python
print('Коэффициент корреляции между жилой площадью и общей площадью квартиры {:.2f}'
      .format(df['living_area'].corr(df['total_area'])))
```

Полученный коэффициент коррекляции говорит нам о том, что жилая площадь и общая площадь квартиры очень тестно взаимосвязаны.  
Найдем усредненный коэффициент отношения жилой площади к общей, и применим его для расчета отсутствующих значений.

```python
# Расчет медианного отношения жилой площади к общей площади квартиры
area_cor_liv = ((df.loc[df['living_area'].isna() == False, 'living_area'])/
     (df.loc[df['living_area'].isna() == False, 'total_area'])).median()

# Применение полученного значения для заполнения пропусков, и округления значения до 2 ед.
df['living_area'] = df['living_area'].fillna((df['total_area']) * area_cor_liv).round(decimals = 2)
```

Пропуски в столбце is_apartment (апартаменты) будут заменены на значения False из соображений, что в случае, если недвижимость является аппартаментами, это бы указали в обязательном порядке, а в остальных случаях этот пункт могли бы пропустить при заполнении объявления.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

Столбец kitchen_area содержит информацию о площади кухни в квадратных метрах (м²). В нем пропущено 2278 значений (или 9,6% от общего объема данных).  

Проверим корреляцию между площадью кухни и общей площадью квартиры.

```python
print('Коэффициент корреляции между площадью кухни и общей площадью квартиры {:.2f}'
      .format(df['kitchen_area'].corr(df['total_area'])))
```

Зависимость присутствует, но куда менее выраженная, чем между жилой и общей площадью.  
Проверим, у какого количества квартир-студий отсутствует информация о площади кухонь.

```python
print('Из', len(df[(df['studio'] == True)]), 'квартир-студий, у', 
      len(df[(df['studio'] == True) & (df['kitchen_area'].isna())]), 'отсутствует площадь кухни.')
```

Квартиры-студии не имеют площадь кухни, что является нормой.

Для заполнения пустых значений площади кухни у не студий посчитаем разницу между общей площадью и жилой площадью, и умножим это значение на коэффициент, отражающий отношение площади кухни ко всей нежилой площади квартиры (т.к. помимо кухни, к нежилым площадям также относятся коридоры, санузлы и тд)

```python
area_cor_kit = ((df.loc[df['kitchen_area'].isna() == False, 'kitchen_area'])/
                (df.loc[df['kitchen_area'].isna() == False, 'total_area'] - 
                 df.loc[df['kitchen_area'].isna() == False, 'living_area'])).median()
df['kitchen_area'] = df.loc[df['studio'] != True, 
                            'kitchen_area'].fillna((df['total_area']-
                                                    df['living_area']) * area_cor_kit).round(decimals = 2)
```

Пропуски в balcony (число балконов) будут заполняться по аналогии с is_apartment.

```python
df['balcony'] = df['balcony'].fillna(False)
```

## Расчёты и добавление результатов в таблицу

Далее нам потребуется добавить в таблицу дополнительные столбцы:  
цену квадратного метра (price_per_sq_m);  
день недели (weekday), месяц (month) и год публикации объявления (year);  
этаж квартиры; варианты — первый, последний, другой (floor_category);  
соотношение жилой и общей площади (live_total_ratio), а также отношение площади кухни к общей (kitchen_total_ratio).

```python
df['price_per_sq_m'] = (df['last_price']/df['total_area']).round().astype('int64')
weekday = pd.DatetimeIndex(df['first_day_exposition'])
df['weekday'] = weekday.day_name()
month = pd.DatetimeIndex(df['first_day_exposition'])
df['month'] = month.month_name()
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

```python
def floor_categorize(data):
    if data['floor'] == 1:
        return 'первый'
    if data['floor'] == data['floors_total']:
        return 'последний'
    return 'другой'

df['floor_category'] = df.apply(floor_categorize, axis=1)

df['live_total_ratio'] = (df['living_area']/df['total_area']).round(decimals = 2)
df['kitchen_total_ratio'] = (df['kitchen_area']/df['total_area']).round(decimals = 2)
```

```python
# Добавляем дополнительные столбцы в таблицы и объединяем их, чтобы сформировать общую гистограмму
df_filtered_spb['name'] = 'Санкт-Петербург'
df_filtered_spb_8km['name'] = 'Сегмент'
df_filtered_spb_for_plot=pd.concat([df_filtered_spb,df_filtered_spb_8km])
sns.barplot(x='floor_category', y='last_price', data=df_filtered_spb_for_plot, hue='name',estimator=np.median)
plt.xlabel('Этаж')
plt.ylabel('Цена, руб.')
plt.show()
```

Благодаря полученной выше гистограмме видна намного более ярко выраженная зависимость стоимости квартиры от этажа для нашего сегмента.  
Для квартир в центре Санкт-Петербурга первый этаж будет куда более важным фактором в оценке стоимости квартиры, нежели чем для города в целом. Квартира на последнем этаже в центре города будет оценена почти также, как и в середине дома. Но эта тенденция характерна и для всего Санкт-Петербурга.

```python
def barplot_median_spb(column, xlabel, ylabel):
    sns.barplot(x=column, y='last_price',data=df_filtered_spb_for_plot, hue='name',estimator=np.median)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
barplot_median_spb('weekday', 'День недели', 'Цена, руб.')
barplot_median_spb('month', 'Месяц', 'Цена, руб.')
barplot_median_spb('year', 'Год', 'Цена, 10 млн. руб.')
```

Стоимость квартир в зависимости от даты размещения для квартир сегмента и квартир всего Санкт-Петербурга схожа. Среди выбивающихся аномалий - медианное значение стоимости квартир в декабре для центрального района города ниже, чем в течение всего года, а для квартир Санкт-Петербурга это не характерно.
